```python
from glob import glob
import os

path = './Artificial-Intelligence/AI_reboot/Keras/Datas/RNN'
txt_path = glob(os.path.join(path, '*.txt'))

txt_path
```

```python
text = [open(txt, 'r', encoding='UTF-8', errors='ignore').read() for txt in txt_path]
text = '\n'.join(text)

text = text.replace('\r\n', ' ')
text = text.replace('\n', ' ')
text = text.replace('\ufeff', '')
text = text.replace('\uf8f7', '')

text
```

```python
chars = list(sorted(set(text)))
char_2_idx = {ch: idx for idx, ch in enumerate(chars)}

char_2_idx
```

```python
from keras.layers import Dense, Input, LSTM, TimeDistributed, Dropout
from keras.optimizers import RMSprop
from keras.models import Model

import tensorflow as tf
from keras.backend import tensorflow_backend as K

config = tf.ConfigProto()
config.gpu_options.allow_growth = True
sess = tf.Session(config=config)
K.set_session(sess)

def char_rnn_model(num_chars, num_layers, nodes=512, drop=0.1):
    i = Input(shape = (None, num_chars), name='input')
    h = LSTM(nodes, return_sequences=True)(i)
    
    for rep in range(num_layers-1):
        h = LSTM(nodes, return_sequences=True)(h)
        h = Dropout(drop)(h)
    o = TimeDistributed(Dense(num_chars, name='dense', activation='softmax'))(h)
    model = Model(i, o)
    opti = RMSprop(lr=0.01)
    model.compile(optimizer=opti, loss='categorical_crossentropy', metrics=['accuracy'])
    return model

model = char_rnn_model(len(chars), num_layers=3, nodes=200, drop=0.3)
model.summary()
```

```python
import numpy as np
import random

def data_generator(all_text, num_chars, CHUNK, batch):
    x = np.zeros((batch, CHUNK, num_chars))
    y = np.zeros((batch, CHUNK, num_chars))
    
    while True:
        for row in range(batch):
            idx = random.randrange(len(all_text) - CHUNK - 1)
            chunk = np.zeros((CHUNK+1, num_chars))
            
            for rep in range(CHUNK+1):
                chunk[rep, char_2_idx[all_text[idx+rep]]] = 1
            x[row, :, :] = chunk[:CHUNK]
            y[row, :, :] = chunk[1:]
        yield x, y

next(data_generator(text, len(char_2_idx), CHUNK = 160, batch=4))
```

```python
chunk = 160
wpath = './Artificial-Intelligence/AI_reboot/Keras/weights/deadpool.h5'
from keras import callbacks
reduce = callbacks.ReduceLROnPlateau(monitor = 'acc', factor = 0.01, patience=10)
check = callbacks.ModelCheckpoint(monitor = 'acc', filepath=wpath, save_best_only=True)
call_lis = [reduce, check]

model.fit_generator(
    data_generator(text, len(chars), batch=150, CHUNK=chunk), epochs = 20,
    callbacks = call_lis ,steps_per_epoch=2*len(text) / (256*chunk))
```

```python
import sys

def generate_output(model, training_text, start_index=None, diversity=None, amount=200):
    if start_index is None:
        start_index = random.randint(0, len(training_text) - chunk - 1)
    generated = training_text[start_index: start_index + chunk]
    yield generated + '#'
    for i in range(amount):
        x = np.zeros((1, len(generated), len(chars)))
        for t, char in enumerate(generated):
            x[0, t, char_2_idx[char]] = 1.
        preds = model.predict(x, verbose=0)[0]
        if diversity is None:
            next_index = np.argmax(preds[len(generated) - 1])
        else:
            preds = np.asarray(preds[len(generated) - 1]).astype('float64')
            preds = np.log(preds) / diversity
            exp_preds = np.exp(preds)
            preds = exp_preds / np.sum(exp_preds)
            probas = np.random.multinomial(1, preds, 1)
            next_index = np.argmax(probas)     
        next_char = chars[next_index]
        yield next_char

        generated += next_char
    return generated

for ch in generate_output(model, text):
    sys.stdout.write(ch)
    
print()
```
