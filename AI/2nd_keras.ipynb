```python
# -*- coding : utf-8 -*-
%matplotlib inline

from keras.layers import Dense, Activation
from keras.models import Sequential
from keras.datasets import mnist
from keras.utils import np_utils
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set()
plt.style.use('ggplot')
np.random.seed(3)
```

```python
# NOTE : Dataset 준비
(train_val, train_lab), (test_val, test_lab) = mnist.load_data()
print(train_val[50000:].shape)

# NOTE 
# train set은 28 x 28사이즈로 60000장의 이미지

print(test_val.shape)

# NOTE
# tets set은 28 x 28사이즈로 60000장의 이미지

# train set과 validation set 분리
vali_val = train_val[50000:]
vali_lab = train_lab[50000:]

train_val = train_val[:50000]
train_lab = train_lab[:50000]

test_img[0].shape
```

```python
# NOTE : 정규화(Normalization)
train_val = train_val.reshape(50000, 784).astype('float32') / 255.0
vali_val = vali_val.reshape(10000, 784).astype('float32') / 255.0
test_val = test_val.reshape(10000, 784).astype('float32') / 255.0
```

```python
# NOTE 
# train set, validation set 고르기

# NOTE : 50000개의 데이터 중 랜덤하게 700개 고름.
train_rand = np.random.choice(50000, 700)
# NOTE : 10000개의 데이터 중 랜덤하게 300개 고름.
vali_rand = np.random.choice(10000, 300)

# NOTE
# 팬시 인덱싱을 활용하여 train set의 데이터 중 랜덤하게 700개의 데이터를 고름.
train_val = train_val[train_rand]
train_lab = train_lab[train_rand]

# NOTE
# 팬시 인덱싱을 활용하여 validation set의 데이터 중 랜덤하게 300개의 데이터를 고름.
vali_val = vali_val[vali_rand]
vali_lab = vali_lab[vali_rand]
```

```python
# NOTE : 라벨링 전환
train_lab = np_utils.to_categorical(train_lab)
vali_lab = np_utils.to_categorical(vali_lab)
test_lab = np_utils.to_categorical(test_lab)
```

```python
# NOTE : 모델 구성
model = Sequential()
model.add(Dense(units = 2, input_dim = 28*28, activation='relu'))
model.add(Dense(units = 10, activation = 'softmax'))

# NOTE 
# unit : 한 층에서 사용할 노드의 개수

# NOTE : 모델 컴파일
model.compile(optimizer = 'adam', loss = 'categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# NOTE : 모델 학습
hist = model.fit(train_val, train_lab, batch_size=10, epochs = 1000, validation_data=(vali_val, vali_lab))

# NOTE : 학습과정 시각화
fig, loss_ax = plt.subplots()
accu_ax = loss_ax.twinx()
fig.set_size_inches(20, 15)


# NOTE
# train set의 loss는 노란색으로 표시
# validation set의 loss는 빨강색으로 표시

loss_ax.plot(hist.history['loss'], 'y', label = 'train loss')
loss_ax.plot(hist.history['val_loss'], 'r', label = 'validation loss')

# NOTE
# train set의 accu는 파란색으로 표시
# validation set의 accu는 초록색으로 표시

accu_ax.plot(hist.history['acc'], 'b', label = 'train accu')
accu_ax.plot(hist.history['val_acc'], 'g', label = 'validation accu')

loss_ax.set_xlabel('epoch')
loss_ax.set_ylabel('Cost')
accu_ax.set_ylabel('Accuracy')

loss_ax.legend(loc='upper left')
accu_ax.legend(loc='lower left')
```
